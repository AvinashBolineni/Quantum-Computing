```
import qiskit as q
%matplotlib inline
```

# Create a New Quantum Circuit

```
circuit = q.QuantumCircuit(2,2)
circuit.x(0)
circuit.h(1)
circuit.draw(output='mpl')
```

```
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

## Initialize a New simulator

### Statevector simulator

```
statevec_sim = q.Aer.get_backend('statevector_simulator')
statevec_job = q.execute(circuit, backend=statevec_sim).result()
statevec = statevec_job.get_statevector()
```

### Qasm simulator

```
qasm_sim = q.Aer.get_backend('qasm_simulator')
qasm_job = q.execute(circuit, backend=qasm_sim).result()
counts = qasm_job.get_counts()
```

# Visualize the results

```
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
```

## visualize the quibit using the results from the statevector simulator

```
plot_bloch_multivector(statevec)
```

## visualize the results gained in an histogram using the qasm simulator

```
plot_histogram(counts)
```

# A Quibit in Superposition

create a single quibit in a quantum circuit, and make that certain quibit go to superposition and check the statevector

```
qc = q.QuantumCircuit(1,1)
qc.h(0)
# qc.measure(0,0)
qc.draw(output='mpl')
```

```
# initializing the statevector simulator

statevec_sim = q.Aer.get_backend('statevector_simulator')
statevec_job = q.execute(qc, backend=statevec_sim).result()
statevec = statevec_job.get_statevector(qc)

# visualizing the statevector

plot_bloch_multivector(statevec)
```

# Two Entangled Quibits

create two quibits, and make each other entangled and look at the results

```
qc1 = q.QuantumCircuit(2,2)
qc1.h(0)
qc1.cx(0,1)
# qc1.measure(0, 0)
qc1.draw(output='mpl')
```

```
# initializing the statevector simulator

statevec_sim = q.Aer.get_backend('statevector_simulator')
statevec_job = q.execute(qc1, backend=statevec_sim).result()
statevec = statevec_job.get_statevector(qc1)

# visualizing the statevector

plot_bloch_multivector(statevec)
```
